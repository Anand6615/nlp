 Program5: Perform basic frequency analysis on a text
 i. Count the frequency of each word in a given text.
 ii. Visualize the word frequencies using a bar chart

```python
import matplotlib.pyplot as plt
from collections import Counter
import re
```

```python
text = """Rolls-Royce is a British luxury automobile and aerospace company known for its exceptional craftsmanship, engineering excellence, and timeless elegance. Founded in 1906 by Charles Rolls and Henry Royce, the brand has become synonymous with prestige and sophistication. Rolls-Royce cars, such as the Phantom, Ghost, and Cullinan, are handcrafted with meticulous attention to detail, offering unparalleled comfort and cutting-edge technology. The companyâ€™s signature Spirit of Ecstasy hood ornament symbolizes luxury and exclusivity. Beyond automobiles, Rolls-Royce is also a leader in aerospace engineering, manufacturing powerful jet engines for commercial and military aircraft. The brand continues to set the standard for innovation and luxury worldwide."""

words = re.findall(r'\b[a-zA-Z]+\b', text.lower())

# Count word frequencies
word_counts = Counter(words)

# Select top 10 most common words
most_common_words = word_counts.most_common(10)
words, counts = zip(*most_common_words)

# Plot bar chart
plt.figure(figsize=(10, 5))
plt.bar(words, counts, color='skyblue')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 10 Word Frequencies in Text')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
import re

# Sample text
text = text

# Preprocess text: convert to lowercase and extract words
words = re.findall(r'\b[a-zA-Z]+\b', text.lower())

# Count word frequencies
word_counts = Counter(words)
most_common_words = word_counts.most_common(10)
words, counts = zip(*most_common_words)

# Plot with Seaborn
plt.figure(figsize=(10, 5))
sns.barplot(x=list(words), y=list(counts), palette="coolwarm")
plt.xlabel("Words")

plt.ylabel("Frequency")
plt.title("Top 10 Word Frequencies in Text")
plt.xticks(rotation=45)
plt.show()
```

```python
import plotly.io as pio
pio.renderers.default = "browser"
from collections import Counter
import re

# Sample text
text = text

# Preprocess text
words = re.findall(r'\b[a-zA-Z]+\b', text.lower())

# Count word frequencies
word_counts = Counter(words)
most_common_words = word_counts.most_common(10)
df = {"Words": [w[0] for w in most_common_words], "Counts": [w[1] for w in most_common_words]}

# Plot with Plotly
fig = px.bar(df, x="Words", y="Counts", title="Top 10 Word Frequencies in Text", color="Counts")
fig.show()
```
